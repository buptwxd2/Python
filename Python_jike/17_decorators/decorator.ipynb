## decorating a function with parameters
## 装饰有参数的函数

```python
def my_decorator(func):
    def wrapper(message):
        print("wrapper of decorator")
        func(message)
    
    return wrapper

@my_decorator
def greet(message):
    print(message)
```

```python
greet("Yoho")
```

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("wrapper of decorator")
        func(*args, **kwargs)
    
    return wrapper

@my_decorator
def greet(message):
    print(message)
```

```python
greet("Yoho")
```

```python
@my_decorator
def another_func(a, b=2):
    print(a, b)
```

```python
another_func("Hello", "Yoho")
```

## decorator meta parameters
## 装饰器本身的参数

```python
## for example: para: num in the below example. Inspect the following code
```

```python
def repeat(num):
    def my_decorator(func):
        def wrapper(*args, **kwargs):
            for i in range(num):
                print("wrapper of decorator")
                func(*args, **kwargs)
    
        return wrapper
    
    return my_decorator
```

```python
# @repeat(4)
def greet(message):
    print(message)
```

```python
greet = repeat(2)(greet)
```

```python
greet
```

```python
greet("hello world")
```

## keep metada info when using decorator ==> functools.wrap

```python
import functools

def my_decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print('wrapper of decorator')
        func(*args, **kwargs)
    return wrapper
    
@my_decorator
def greet(message):
    print(message)
```

```python
help(greet)
```

```python
class Count:
    def __init__(self, func):
        self.func = func
        self.num_calls = 0

    def __call__(self, *args, **kwargs):
        self.num_calls += 1
        print('num of calls is: {}'.format(self.num_calls))
        return self.func(*args, **kwargs)

@Count
def example():
    print("hello world")
```

## python built_in decorator: lru_cache

```python
from functools import lru_cache
```

```python
import time
```

```python
def time_usage(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print("Time duraton is {}".format(end_time - start_time))
        return result
    
    return wrapper
```

```python
# Fibonacci numbers 
@time_usage
def fib(n):
    
    if n < 2:
        return n
    
    return fib(n-1) + fib(n-2)
```

```python
fib(20)
```

```python
# Fibonacci numbers 
@lru_cache(maxsize=None)
@time_usage
def fib(n):
    
    if n < 2:
        return n
    
    return fib(n-1) + fib(n-2)
```

```python
fib(100)
```
